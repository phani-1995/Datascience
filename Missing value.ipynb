```python
#Alternative way of filling missing value with test statistic is by using our 
#Imputer method found in sklearn.preprocessing

from sklearn.processing import Imputer
imp = Imputer(missing_value='NaN', strategy= 'mean', axis='0')
imp.fit(df)
imp.transform(df)
#For mode we use strategy = 'most_frequent'
```
